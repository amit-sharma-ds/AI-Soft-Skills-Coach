```python
import pandas as pd
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Creating DataFrame from your CSV
df = pd.read_csv(r"C:\Users\Acer\Downloads\archive (15)\dataset.csv")

# Feature extraction using TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['answer'])

# Function to calculate cosine similarity between user answer and dataset answers
def get_similarity(user_answer):
    user_answer_vector = vectorizer.transform([user_answer])
    answer_vectors = vectorizer.transform(df['answer'])
    similarity = cosine_similarity(user_answer_vector, answer_vectors)
    return similarity.max()

# Function to provide graded feedback based on similarity score
def provide_feedback(user_answer):
    # Get similarity score
    similarity_score = get_similarity(user_answer)

    # Generate feedback based on similarity score with improved feedback
    if similarity_score < 0.3:
        feedback = "Your answer needs more depth. Consider adding relevant details or examples that explain key concepts more clearly."
    elif 0.3 <= similarity_score < 0.5:
        feedback = "Your answer is on the right track but could be more comprehensive. Try elaborating on your points with more specific examples."
    elif 0.5 <= similarity_score < 0.7:
        feedback = "Good job! Your answer is relevant. For greater clarity, consider expanding on key concepts and providing more detail."
    else:
        feedback = "Excellent! Your answer is highly relevant and covers all key points. Keep it up and consider adding real-world examples for further depth."

    return feedback

# Function to start the mock interview
def mock_interview():
    # Randomly select a question from the dataframe
    random_index = random.randint(0, len(df) - 1)
    question = df['question'].iloc[random_index]

    # Display the question to the user
    print(f"Question: {question}")

    # Collect user answer
    user_answer = input("Your answer: ")

    # Provide feedback based on the user answer using Cosine Similarity
    feedback = provide_feedback(user_answer)
    print(f"Feedback: {feedback}")

# Start the mock interview
mock_interview()
```
